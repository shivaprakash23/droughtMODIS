```python
import arcpy
from datetime import datetime, timedelta
import re

# Access the current ArcGIS Pro project and active map
aprx = arcpy.mp.ArcGISProject("CURRENT")
m = aprx.activeMap
```

```python
mosaic_layer_name = "VCIModis"
mosaic_layer = m.listLayers(mosaic_layer_name)[0]
mosaic_dataset = mosaic_layer.dataSource
```

```python
mosaic_layer
```

```python
field_name = "AcqDate"
if field_name not in [f.name for f in arcpy.ListFields(mosaic_dataset)]:
    arcpy.AddField_management(mosaic_dataset, field_name, "DATE")
```

```python
with arcpy.da.UpdateCursor(mosaic_dataset, ["Name", field_name]) as cursor:
    for row in cursor:
        name = row[0]
        match = re.search(r"A(\d{4})(\d{3})", name)
        if match:
            year = int(match.group(1))
            doy = int(match.group(2))
            date_val = datetime(year, 1, 1) + timedelta(days=doy - 1)
            row[1] = date_val
            cursor.updateRow(row)

print("✅ All acquisition dates successfully extracted to 'AcqDate' field in VCIModis.")
```
